```python
#merging and Concatenating
```

```python
df1 = pd.DataFrame({
    'Item': ['A', 'B', 'C'],
    'Price': [10,20,30]
})

df2 = pd.DataFrame({
    'Item': ['D', 'E', 'F'],
     'Price': [40,50,60]
})
```

```python
df1
```

```python
pd.concat([df1, df2]).reset_index().drop('index', axis = 1)
```

```python
df3 = pd.DataFrame({
    'Item': ['A', 'B', 'C'],
    'Price': [10,20,30] 
})

df4 = pd.DataFrame({
    'Country': ['X', 'Y', 'Z'],
    'Available': [True, True, False]
})
```

```python
pd.concat([df3, df4], axis = 1)
```

```python
df5 = pd.DataFrame({
    'Item': ['B', 'C', 'D'],
    'Country': ['X', 'Y', 'Z']
})
```

```python
pd.merge(df3, df5,on='Item' how='right')
```
